```python
# Imports
import math

from plot import plot_rho_vs_M, plot_kstar_vs_M
import pandas as pd
from experiment import bound_sufficiency_df, is_N_enough
```

```python
N = 200
alpha = (math.sqrt(5) - 1) / 2  # golden ratio conjugate ~0.618

df = bound_sufficiency_df(
    N=N,
    alpha=alpha,
    multipliers=(1, 2, 4, 8),
    k2_window=3,
    method="numpy"
)

df
```

Observe that rho, k1_star, k2_star stay the exact same as M increases
Thus we can observe that this is evidence for -N, N being enough for this N, alpha

```python
# Compares the result at M = N vs M = 2N
print("Is M=N enough?", is_N_enough(df))
```

```python
# rho vs search size M
plot_rho_vs_M(df, title=f"rho vs M (N={N}, alphaâ‰ˆ{alpha:.6f})")
```

Since flat line it looks like -N, N is enough

```python
# Minimizing vector k* vs M
plot_kstar_vs_M(df, title=f"k* vs M (N={N}, alphaâ‰ˆ{alpha:.6f})")
```

Thus....

```python
# Already shown for N vs @N now used for all multipliers
def is_enough_strict(df, tol=1e-9):
    base = df[df["mult"] == 1].iloc[0]
    for _, row in df.iterrows():
        same_rho = abs(row["rho"] - base["rho"]) <= tol * max(1.0, abs(base["rho"]))
        same_k = (row["k1_star"] == base["k1_star"]) and (row["k2_star"] == base["k2_star"])
        if not (same_rho and same_k):
            return False
    return True

print("Strict enough (stable across all M tested)?", is_enough_strict(df))
```

To make it even more interesting we will test lots of alphas and ns

```python
# Picks interesting alphas, Ns
alphas = {
    "golden": (math.sqrt(5) - 1) / 2,
    "sqrt2": math.sqrt(2) % 1,
    "e": math.e % 1,
}
Ns = [50, 100, 200, 400]

rows = []
for name, a in alphas.items():
    for N in Ns:
        df = bound_sufficiency_df(N, a, multipliers=(1,2,4), k2_window=3, method="numpy")
        rows.append({
            "alpha_name": name,
            "alpha": a,
            "N": N,
            "enough_simple": is_N_enough(df),
            "enough_strict": is_enough_strict(df),
            "rho_at_N": float(df[df["mult"]==1]["rho"].iloc[0]),
            "k_star_at_N": (
                int(df[df["mult"]==1]["k1_star"].iloc[0]),
                int(df[df["mult"]==1]["k2_star"].iloc[0]),
            )
        })

summary = pd.DataFrame(rows)
summary
```

```python
# Visualize Pass vs Fail
import matplotlib.pyplot as plt

plt.figure()
for name in summary["alpha_name"].unique():
    s = summary[summary["alpha_name"] == name].sort_values("N")
    plt.plot(s["N"], s["enough_strict"].astype(int), marker="o", label=name)

plt.xlabel("N")
plt.ylabel("enough_strict (1=yes, 0=no)")
plt.title("Is [-N,N]^2 enough? (empirical)")
plt.ylim(-0.1, 1.1)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
